#### Lets do some vector operations

```python
#what if we wanted to add two different elements from two different lists?
def add_vector_components(v, w):
    """add corresponding elements in vectors(lists)"""
    return [v_i + w_i for v_i, w_i in zip(v,w)]

def subtract_vector_components(v, w):
    """subtract corresponding elements in vectors(lists)"""
    return [ v_i - w_i for v_i, w_i in zip(v,w)]

#What if we want to run a reduce function where we collapse/sum all the values of each component in a list of vectors to a single vector

def vector_sum(vectors):
    """sums all corresponding elements"""
    result = vectors[0] #use this as our beginning position
    for vector in vectors[1:]:
        result = add_vector_components(result, vector) #go through each vector in the list and sum their corresponding components
    return result

#we can also just call a reduce function

def vector_sum_reduce(vectors):
    return reduce(add_vector_components, vectors) # reduce function applies whichever function we call to each component in the vector


#what if we want to scale our vector? We need to multiply each component by some value.

def scalar_multiply(scalar, vector):
    return [scalar * vector_i for vector_i in vector]


#what if we want to get a component wise means for a list of same-sized vectors

def vector_mean(vectors):
    """compute the vector whose ith element is the mean of the ith elements of the input vectors"""
    
    #lets get our denominator
    n = len(vectors)
    #multiply each by 1/n to get the mean of the vector
    return scalar_multiply(1/n, vector_sum_reduce(vectors))

#dot product of two vectors is the sum of their component wise products

def dot_product(v,w):
    """v_1 * w_1 + ... + v_n * w_n"""
    
    return sum(v_i * w_i for v_i, w_i in zip(v,w))

#Now that we have dot product we can do a sum of squares for each vector

def sum_of_squares(v):
    """v_1*v_1 + ... + v_n*v_n"""
    
    return dot_product(v,v)

#now we can compute the length of a vector

#we need to import the math module since we'll be doing square root

import math

def length_magnitude(v):
    return math.sqrt(sum_of_squares(v))


def squared_distance(v, w):
    """(v_1 - w_1)**2 + ... + (v_n - w_n)**2"""
    
    return sum_of_squares(subtract_vector_components(v, w))

#now we have a distance function (pythagoras)

def distance(v, w):
    
    return math.sqrt(squared_distance(v,w))


#alternatively

def distance_alt(v,w):
    
    return length_magnitude(subtract_vector_components(v,w)) 
```

#### Lets look at some matrix operations

```python
#lets find out the shape of our matrix
#in Python matrices are represented as lists of lists
def shape(A):
    num_rows = len(A)
    num_cols = len(A[0]) if A else 0
    return num_rows, num_cols

#lets get the ith row of a matrix

def get_row(A, i):
    return A[i]

#lets get the jth column of a matrix

def get_col(A, j):
    
    return [A_i[j]       #jth element in row i of matrix A
            for A_i in A] #for each row i in A

#Lets construct a matrix with  i rows and j columns using some function to create values

def make_matrix(num_rows, num_cols, generator_fn):
    """returns a num_rows X num_cols matrix
    whose (i,j)th entry is generator_fn(i,j)"""
    
    return [[generator_fn(i, j)         #given i, create a list
             for j in range(num_cols) ] # [generator_fn(i ,0), ... generator_fn(i, 1)...]
            for i in range(num_rows)] #create a list for each i

#we can use this to make an identity function

def diagonal_id_matrix(i, j):
    """1's on each diagonal place, 0 everywhere else"""
    
    return 1 if  i == j else 0 



```
