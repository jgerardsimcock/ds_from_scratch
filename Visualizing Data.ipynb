```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda grade: grade // 10 * 10 

def make_histogram(*args):

    histogram = Counter(decile(grade) for grade in grades)


    plt.bar([x - 4 for x in histogram.keys()], #this will shift the bar left 4 
            histogram.values(),                #assign height of bar to # of students with that score
            8)                                 #each bar is width 8

    plt.axis([-5, 105, 0, 5])      #x-axis from -5 to 105, y-axis from 0 to 5
    
    plt.xticks([10 * i for i in range(11)]) #put a label at 0, 10, ...100
    
    plt.xlabel("Decile")
    plt.ylabel("# of students")
    plt.title("Distribution of Exam 1 Grades")
    plt.show()
    

make_histogram(plt, grades, decile)
```

```python
def show_misleading_chart(plt, mislead=True):

    mentions = [500, 505]
    years = [2013, 2014]

    plt.bar([2012.6, 2013.6], mentions, 0.8)
    plt.xticks(years)
    plt.ylabel("# of times a surfer says stoked per day")

    plt.ticklabel_format(useOffset=False)
    
    if mislead:
        plt.axis([2012.5, 2014.5, 499, 506])
        plt.title("Look at all that Stoke!")
        
    else:
        plt.axis([2012.5, 2014.5, 0, 550])
        plt.title("Not so stoked")
        
    plt.show()
    
show_misleading_chart(plt, mislead=True)
show_misleading_chart(plt, mislead=False)
```

```python


def show_diff_line_chart(plt):
    
    variance     = [1,2,4,8,16,32,64,128,256]
    bias_squared = [256,128,64,32,16,8,4,2,1]
    
    total_error = [x + y for x, y in zip(variance, bias_squared)]
    
    xs = range(len(variance))
    
    #we are going to call plt.plot many times and layer the lines
    plt.plot(xs, variance, 'g-', label='variance') #green solid
    plt.plot(xs, bias_squared, 'r-.', label='bias^2') #red dashed
    plt.plot(xs, total_error, 'b:', label='total error') #blue dotted

    #as we have labeled our series, we can call plt.legend()
    plt.legend(loc=9)
    plt.xlabel("model complexity")
    plt.title("The Bias-Variance Tradeoff")
    plt.show()
    
show_diff_line_chart(plt)
    
```

```python
def friend_scatter_plot(plt):
    
    friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
    minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
    labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
    
    plt.scatter(friends, minutes)
    
    for label, friend_count, minute_count in zip(labels, friends, minutes):
        plt.annotate(label,
                     xy=(friend_count, minute_count),
                     xytext=(-5, 5),
                     textcoords='offset points')
        
    plt.title("Daily Minutes vs Number of Friends")
    plt.xlabel("#of Friends")
    plt.ylabel("daily minutes spent on site")
    plt.show()
        
        
friend_scatter_plot(plt)
    
```

```python
def compare_scatterplot_axes(plt, equal_axes=False):
    
    test_1_grades = [ 99, 90, 85, 97, 80]
    test_2_grades = [100, 85, 60, 90, 70]
    
    plt.scatter(test_1_grades, test_2_grades)
    
    plt.xlabel("Test 1 Grade")
    plt.ylabel("Test 2 Grade")
    
    if equal_axes:
        plt.title("Axes are Comparable")
        plt.axis("equal")
        
    else:
        plt.title("Axes Aren't Comparable")
        
    plt.show()

    
compare_scatterplot_axes(plt, equal_axes=False)
compare_scatterplot_axes(plt, equal_axes=True)
    
```
